
# 3.BAR GRAPH OF MALE VS FEMALE BUYERS PARTICIPATED IN THE SALES

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
cardata='downloads/cars_data.csv'
df=pd.read_csv(cardata)
plt.ylim(0, 10000)
series = df["Buyer Gender"].value_counts()
series.plot(kind='bar')
plt.title('BAR GRAPH OF MALE AND FEMALE BUYERS')
plt.xlabel('MALES VS FEMALE BUYERS PARTICIPATED IN SALES')
plt.ylabel('NUMBER OF CAR BUYERS')
plt.grid()
plt.show()


```

```python
series
```

INFERENCE:There is a slight increase in number of female buyers.
